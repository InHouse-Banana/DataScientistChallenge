# 1.0 ETL

## 1.1 PAY_1 column is not defined in the data dictionary hence discarded
## PAY_0, and PAY_2 to PAY_6 are dropped because their inclusion would create multiple categories that may have very few elements, hence, possibly, making the model overfit.
## The column ID won't be useful for a prediction so it'll be dropped as well

```python
cols_to_drop = [ 'PAY_' + str(x) for x in range(0, 7)] + ['ID']
```

```python
clean_data = source_data.copy()\
             .drop(columns = cols_to_drop)
display(clean_data.info())
clean_data.head()
```

## 1.2 There are multiple codings for the columns given in the README.md, hence I'll use them to translate the columns.

```python
SEX_dict       = {1 : 'male', 2 : 'female'}
EDUCATION_dict = {1 : 'graduate_school', 2 : 'university', 3: 'high_school'\
                  , 4 : 'education_others', 5 : 'unknown', 6 : 'unknown'}
MARRIAGE_dict  = {1 : 'married', 2 : 'single', 3 : 'marriage_others'}
#default_payment_next_month_dict = {0 : 'no', 1 : 'yes'}
```

```python
clean_data['SEX']       = clean_data['SEX'].replace(SEX_dict)
clean_data['EDUCATION'] = clean_data['EDUCATION'].replace(EDUCATION_dict)
clean_data['MARRIAGE']  = clean_data['MARRIAGE'].replace(MARRIAGE_dict)
#clean_data['default.payment.next.month']  = clean_data['default.payment.next.month'].replace(default_payment_next_month_dict)

clean_data.head()
```

### Let's check that the replacement worked

```python
for col in ['SEX', 'EDUCATION', 'MARRIAGE']:
    display(clean_data[col].value_counts())
```

### There are records that do not appear in the data dictionary. 
### While these values might be grouped under others or unknown, unless this is confirmed the best practice is to ignore them

```python
clean_data = clean_data[clean_data['EDUCATION'] != 0]
clean_data  = clean_data[clean_data['MARRIAGE'] != 0]
```

### Let's check that the replacement worked

```python
for col in ['SEX', 'EDUCATION', 'MARRIAGE', 'default.payment.next.month']:
    display(clean_data[col].value_counts())
```

```python
display(clean_data.info())
clean_data
```

## 1.3 Since the importance value is how much of a Bill a person paid, I'll define a BAL_AMT = BILL_AMT - PAY_AMT, and drop BILL_AMT, PAY_AMT. 
## Having this feature should capture the same information and also making the model training faster by having one less feature to train on.

```python
for month in range(1,7):
    bill_col = 'BILL_AMT' + str(month) 
    pay_col  = 'PAY_AMT' + str(month) 
    bal_col  = 'BAL_AMT' + str(month) 
    clean_data[bal_col] = clean_data[bill_col] - clean_data[pay_col]
    clean_data = clean_data.drop(columns = [bill_col, pay_col])
```

```python
display(clean_data.info())
clean_data.head()
```

# 4.0 One-hot encoding for categorical features

```python
cat_features       = training_data.select_dtypes(include=["object"]).columns.values
print('cat_features:', cat_features)


numerical_features = ['LIMIT_BAL', 'AGE', 'BAL_AMT1'\
                     ,'BAL_AMT2', 'BAL_AMT3', 'BAL_AMT4', 'BAL_AMT5', 'BAL_AMT6']
print('numerical_features:', numerical_features)

target             = ['default.payment.next.month']
print('target:', target)
```

```python
x = training_data[list(cat_features) + list(numerical_features)]
display(x.head())
y = training_data[target]
display(y.head())
```

```python
enc      = OneHotEncoder(handle_unknown='ignore')
x_onehot = enc.fit_transform(x[cat_features]).toarray()

enc_categories = list(itertools.chain(*np.array(enc.categories_, dtype=object)))
x_onehot       = pd.DataFrame(x_onehot, columns = enc_categories).astype(str)

display(x_onehot.info())
x_onehot.head()
```

```python
features_targets         = pd.concat([x_onehot.reset_index(drop=True)
                                      , x[numerical_features].reset_index(drop=True)
                                      , y.reset_index(drop=True)]
                                     , axis = 1
                                    )
features_targets.columns = features_targets.columns.str.lower()
display(features_targets.info())
features_targets.head()
```

### 4.1 Save the one-hot encoded features and targets

```python
features_targets.to_csv('data_artifacts/training_features_targets.csv', index=False)
```

```python
import pickle
with open("model_artifacts/encoder.pkl", "wb") as f: 
    pickle.dump(enc, f)
```
